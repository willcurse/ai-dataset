# 5: How does the global expectation for AI adoption correlate with AI software revenue?

```python
data.head()
```

```python

adoption_expectation = data['Global Expectation for AI Adoption (%)']
software_revenue = data['AI Software Revenue(in Billions)']


correlation = adoption_expectation.corr(software_revenue)

print(f'The correlation between global AI adoption expectation and AI software revenue is: {correlation}')
```

# Interpretation:
A positive correlation would suggest that as the global expectation for AI adoption rises, AI software revenue tends to increase as well.

A negative correlation would suggest that higher AI adoption expectations lead to lower AI software revenue.

A correlation near 0 indicates little to no linear relationship between the two variables.
